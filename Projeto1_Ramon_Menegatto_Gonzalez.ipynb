## Introdução

**Organização: SADC(Southern African Development Community).** 
 
 - A SADC é um bloco econômico qeu possui como objetivo proporcionar o crescimento das economias dos países africanos e consequentemente, melhorar a qualidade de vida de suas populações.
 - A SADC atualmente é composta por 14 países localizados no sul do continente Africano.

## Perguntas a serem respondidas


- O crecimento econômico desse bloco pode ser comparado proporcionalmente com outros países ou organizações?
- A expectativa de vida da população aumentou quando comparados aos outros países da África?


## Explicação da hipótese e mecanismo

A SADC é um bloco econômico dos países do sul da África que se organizam e planejam maneiras de melhorar economicamente e politicamente. Por isso é de se esperar que após alguns anos de sua criação esses países devem ter se desenvolvidos de uma maneira mais acelerada e mais eficiente que os demias países deste continente. Também espero ver uma evolução nos países que entraram nesse bloco econômico. 


Devido à esses acordos ecônomicos e políticos realizados entre os países da SADC é de se experar que a qualidade de vida da população tenha melhorado, tanto em relação a educação, quanto hospitalar e por consequência a expectativa de vida deverá aumentar.

Para avaliar essas hipóteses e responder essas perguntas, será necessário avaliar a renda per capita dos países pertencentes a esse bloco e compará-las a outros países africanos e de outros continentes. Também será necessário avaliar os dados relacionados à expectativa de vida de anos anteriores e compara-los a dados mais atuais.
